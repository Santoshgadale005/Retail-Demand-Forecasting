## 📁 Folder Structure Verification

````python
from pathlib import Path

project_root = Path('..').resolve()
print(f"Project Root: {project_root}\n")

required_dirs = [
    'data/raw', 'data/processed', 'data/external',
    'notebooks', 'warehouse', 'dbt_project',
    'forecasting/prophet', 'forecasting/arima', 'forecasting/lightgbm',
    'inventory', 'streamlit', 'reports',
    'dashboards', 'deployment', 'docker',
    'docs', 'tests', 'configs',
]

print(f"{'Directory':<40} {'Status'}")
print('=' * 55)
for d in required_dirs:
    exists = (project_root / d).exists()
    status = '✅ Found' if exists else '❌ Missing'
    print(f"{d + '/':<40} {status}")
````

## 🔧 Environment Variables

````python
import os
from dotenv import load_dotenv

load_dotenv(project_root / '.env')

env_vars = [
    'GCP_PROJECT_ID', 'BQ_DATASET', 'BQ_LOCATION',
    'GCP_CREDENTIALS_PATH', 'KAGGLE_USERNAME',
    'APP_ENV', 'LOG_LEVEL', 'FORECAST_HORIZON',
]

print(f"{'Variable':<30} {'Value'}")
print('=' * 60)
for var in env_vars:
    value = os.getenv(var, 'NOT SET')
    # Mask sensitive values
    if 'KEY' in var or 'CREDENTIALS' in var:
        display = value[:5] + '...' if len(value) > 5 else value
    else:
        display = value
    print(f"{var:<30} {display}")
````

## ☁️ BigQuery Connection Test

````python
try:
    from google.cloud import bigquery
    
    project_id = os.getenv('GCP_PROJECT_ID')
    client = bigquery.Client(project=project_id)
    
    # List datasets
    datasets = list(client.list_datasets())
    print(f"Connected to BigQuery project: {project_id}")
    print(f"Available datasets: {len(datasets)}")
    for ds in datasets:
        print(f"  📂 {ds.dataset_id}")
        
except Exception as e:
    print(f"BigQuery connection test: {e}")
    print("This is expected if GCP credentials are not yet configured.")
    print("Run: gcloud auth application-default login")
````
